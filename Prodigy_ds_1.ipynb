```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("2k15.csv")
```

# # Histogram of Happiness Score

```python
plt.hist(df['Happiness Score'], bins=10, edgecolor='black')
plt.xlabel('Happiness Score')
plt.ylabel('Frequency')
plt.title('Distribution of Happiness Score')
plt.show()
```

# Happiness Score Vs Family,Freedom,Generosity

```python
df.plot(x="Happiness Score", y=["Family"])
plt.show()
```

```python
df.plot(x="Happiness Score", y=["Freedom"],color='Orange')
```

```python
df.plot(x="Happiness Score", y=["Generosity"],color="Black")
```

# Pieplot of Region and Happiness Score 

```python
texts=df.groupby(['Region']).sum().plot(kind='pie', y='Happiness Score')
```

# # Bar chart: Happiness Rank by Country


```python
plt.figure(figsize=(10, 35))
df_sorted = df.sort_values('Happiness Rank', ascending=True)
plt.barh(df_sorted['Country'], df_sorted['Happiness Rank'])
plt.xlabel('Country')
plt.ylabel('Happiness Rank')
plt.title('Happiness Rank by Country')
plt.show()
```

# How does the Happiness Score vary across different regions?

```python
df.groupby('Region')['Happiness Score'].mean().plot(kind='bar')
plt.xlabel('Region')
plt.ylabel('Average Happiness Score')
plt.title('Average Happiness Score by Region')
plt.show()
```

```python
plt.hist(df['Family'],bins=10, edgecolor='black')
xlabel='Family'
ylabel='Freguency'
plt.show()
```

 # Scatter plot: Happiness Score vs. Economy GDP per Capita


```python
plt.scatter(df['Economy (GDP per Capita)'], df['Happiness Score'])
plt.xlabel('Economy (GDP per Capita)')
plt.ylabel('happiness')
plt.title('Happiness Score vs. Economy GDP per Capita')
plt.show()
```
